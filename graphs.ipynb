```python
import matplotlib.pyplot as plt
import math

count4 = 0
meanH = 0
meanHT = 0
meanL = 0
meanLT = 0

fileName = "MasterExperiment_PC.csv"

with open(fileName, "r") as data:
    
    for line in data:
        numbers = line.split(',')
        if numbers[0].isnumeric():
            if numbers[8] == "edge":
                count4 += 1
                meanH += float(numbers[5])
                meanHT += float(numbers[6])
                meanL += float(numbers[3])
                meanLT += float(numbers[4])

meanH = meanH/count4
meanHT = meanHT/count4
meanL = meanL/count4
meanLT = meanLT/count4


print(count4, meanL, meanLT, meanH, meanHT)

```

```python
def stand_d(avg, pos, fileName):
    file = open(fileName, "r")
    std_de = 0
    count = 0
    for line in file:
        vals = line.split(',')
        if vals[8] == "edge":
            distance = (float(vals[pos]) - avg)**2
            std_de += distance
            count += 1
    std_de /= count
    std_de = math.sqrt(std_de)

    file.close()

    return std_de


stdH = stand_d(meanH, 5, fileName)
stdHT = stand_d(meanHT, 6, fileName)
stdL = stand_d(meanL, 3, fileName)
stdLT = stand_d(meanLT, 4, fileName)


print(stdH, stdHT, stdL, stdLT)
```

```python

l = []
lt = []
h = []
ht = []
z = []

with open(fileName, "r") as data:
    
    for line in data:
        numbers = line.split(',')
        if numbers[0].isnumeric():
            if numbers[8] == "edge":
                l.append((float(numbers[3]), 1))
                lt.append((float(numbers[4]), 1))
                h.append((float(numbers[5]), 1))
                ht.append((float(numbers[6]), 1))
                z.append((float(numbers[7]), 1))

            if numbers[8] == "repeating" or numbers[8] == "dead":
                l.append((float(numbers[3]), 0))
                lt.append((float(numbers[4]), 0))
                h.append((float(numbers[5]), 0))
                ht.append((float(numbers[6]), 0))
                z.append((float(numbers[7]), 0))

            if numbers[8] == "chaos":
                l.append((float(numbers[3]), 2))
                lt.append((float(numbers[4]), 2))
                h.append((float(numbers[5]), 2))
                ht.append((float(numbers[6]), 2))
                z.append((float(numbers[7]), 2))

print(lt)
```

```python
plt.scatter(*zip(*l))
plt.xlabel("Lambda")
plt.ylabel("Class")
plt.savefig("Lambda.jpg")
plt.show()

plt.scatter(*zip(*lt))
plt.xlabel("Lambda_T")
plt.ylabel("Class")
plt.savefig("Lambda_T.jpg")
plt.show()

plt.scatter(*zip(*h))
plt.xlabel("H")
plt.ylabel("Class")
plt.savefig("H.jpg")
plt.show()

plt.scatter(*zip(*ht))
plt.xlabel("H_T")
plt.ylabel("Class")
plt.savefig("H_T.jpg")
plt.show()

plt.scatter(*zip(*z))
plt.xlabel("Z")
plt.ylabel("Class")
plt.savefig("Z.jpg")
plt.show()
```
